
Are there more boys' names or girls' names? What about for particular first letters? What about for ''every'' first letter?


```python
if len(ssadata.boys) > len(ssadata.girls):
    print("There are more boys' names.")
else:
    print("There are more girls' names.")
    
count_b = 0
count_g = 0
for name in ssadata.boys.keys():
    if "a" in name[0]:
        count_b = count_b + 1
for g_name in ssadata.girls.keys():
    if "a" in g_name[0]:
        count_g = count_g + 1
print("There are " + str(count_b + count_g) + " A letter names.")

letters = "abcdefghijklmnopqrstuvwxyz"

for letter in letters:
    count_b = 0
    count_g = 0
    for name in ssadata.boys.keys():
        if letter in name[0]:
            count_b = count_b + 1
    for g_name in ssadata.girls.keys():
        if letter in g_name[0]:
            count_g = count_g + 1
    print("There are " + str(count_b + count_g) + " " + letter.upper() + " names.")
```
